```python
import numpy as np

import pandas as pd
 
import matplotlib.pyplot as plt
 
import statsmodels.api as sm
 
import seaborn
 
seaborn.set()
```

```python
x1= df[['Cement']]
y=df[['Strength']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```
